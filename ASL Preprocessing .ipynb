```python
def kane(row, fill):
    width=0
    leftmax=0
    leftupdated=False
    rightmax=0
    rightupdated=False
    i=10
    j=190
    while i<200 and j>=0:
        if fill[row][i]==0 and leftupdated==False:
            leftmax=i
            #print(leftmax)
            leftupdated=True
        if fill[row][j]==0 and rightupdated==False:
            rightmax=j
            #print(rightmax)
            rightupdated=True
        i+=1
        j-=1
    width=abs(rightmax-leftmax)
    return width
        
        
```
